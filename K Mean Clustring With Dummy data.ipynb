### Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
%matplotlib inline
```

### Import Dataset

```python
from sklearn.datasets import make_blobs
```

### makig object for data

```python
# n_features=2 is basically the number of columns
data = make_blobs(n_samples=200, n_features=2, centers=4, cluster_std=1.8,
                  random_state=101)
```

```python
type(data[0])
```

```python
# labels 
data[1]
```

### Visualize Data

```python
# make a graph between both rows of give samples
sn.set_style('darkgrid')
plt.scatter(data[0][:,0], data[0][:,1], c = data[1], cmap = 'rainbow')
```

### Creat Clusters

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=4)
```

```python
model.fit(data[0])
```

```python
model.cluster_centers_
```

```python
model.labels_
```

### Visualize both the clusters that we made and the lables given in data set

```python
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True,figsize=(10,6))
ax1.set_title('K Means')
ax1.scatter(data[0][:,0],data[0][:,1],c=model.labels_,cmap='rainbow')
ax2.set_title("Original")
ax2.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```
